```python
import tkinter, tkinter.messagebox, tkinter.simpledialog, os ,configparser



class kierunek:
    def __init__(self, x, y):
        self.x=x
        self.y=y
        
class kierunki:
    def __init__(self):
        self.tablica=[]
        self.tablica.append(kierunek(-1,0))
        self.tablica.append(kierunek(-1,1))
        self.tablica.append(kierunek(-1,-1))
        self.tablica.append(kierunek(0,1))
        self.tablica.append(kierunek(0,-1))
        self.tablica.append(kierunek(1,-1))
        self.tablica.append(kierunek(1,0))
        self.tablica.append(kierunek(1,1))

class wierzcholek:
    def __init__(self, x, y):
        self.x=x
        self.y=y
        self.cecha=-1
        self.sciana=0
        self.trasa=0
        
    def DIST(self, sasiad):
        odl=abs(self.x-sasiad.x)+abs(self.y-sasiad.y)
#        print(odl)
        if(odl>1):        
            return 3
        else:
            return 2
    def cechuj(self, sasiad):
        return sasiad.cecha+ self.DIST(sasiad)
        
    def buduj(self):
        self.sciana=1
        
class grupa:
    
    def __init__(self):
        self.tablica=[]

    def nastepniki(self, mapa):
        nastepniki=grupa()
        for  i in range (0, len(self.tablica)):
            kier=kierunki()
            for j in range(0, len(kier.tablica)):
                duzeX=self.tablica[i].x+kier.tablica[j].x
                duzeY=self.tablica[i].y+kier.tablica[j].y
                if((duzeX<0 or duzeX>mapa.duzeX-1 or duzeY<0 or duzeY>mapa.duzeY-1)!=1):
                    skan=mapa.tablica[duzeX][duzeY]
                    if ((skan.cecha==-1) and (skan.sciana!=1)):
                        nastepniki.tablica.append(skan)
            
            
        return nastepniki
    
    def cechujnastepniki(self,mapa):
        for  i in range (0, len(self.tablica)):
            kier=kierunki()
#            print(len(self.tablica))
            for j in range(0, len(kier.tablica)):
                duzeX=self.tablica[i].x+kier.tablica[j].x
                duzeY=self.tablica[i].y+kier.tablica[j].y
#                print(duzeX)
#                print(duzeY)
                if((duzeX<0 or duzeX>mapa.duzeX-1 or duzeY<0 or duzeY>mapa.duzeY-1)!=1):
                    skan=mapa.tablica[duzeX][duzeY]
                    if (skan.cecha!=-1 and skan.sciana!=1):
                        test= self.tablica[i].cechuj(skan)
#                    print(test)
                        if(test<self.tablica[i].cecha or self.tablica[i].cecha==-1):
                            self.tablica[i].cecha=test
        for  i in range (0, len(self.tablica)):
#            print(self.tablica[i].cecha)
            mapa.tablica[self.tablica[i].x][self.tablica[i].y]=self.tablica[i]
        
         
class mapa:
    def __init__(self, x, y):
        self.tablica=[]
        self.duzeX=x
        self.duzeY=y
        self.START=wierzcholek(0,0)
        self.END=wierzcholek(0,0)
        for i in range (0, x):
            wiersz=[]
            for j in range(0, y):
                wiersz.append(wierzcholek(i,j))
            self.tablica.append(wiersz)
            
    def donaldtrump(self, x, y):
        self.tablica[x][y].buduj()
        
    def SetStart(self, wierzcholek):
        self.START.cecha=-1
        self.START=wierzcholek
        wierzcholek.cecha=0
        self.tablica[wierzcholek.x][wierzcholek.y]=wierzcholek
        
    def SetEND(self, wierzcholek):
        self.END=wierzcholek
        
    def CECHUJMAPE(self,buttons):
        poprzedniki=grupa()
        poprzedniki.tablica.append(self.START)
        self.END.cecha==-1
#        print("DADADA")
        while(self.END.cecha==-1):
            nast=poprzedniki.nastepniki(self)
            nast.cechujnastepniki(self)
            poprzedniki=nast
            self.END=self.tablica[self.END.x][self.END.y]
            for x in range(0, self.duzeX):
                for y in range(0, self.duzeY):
                    buttons[x][y].config(text=self.tablica[x][y].cecha)
#            print(self.END.cecha)
#            for i in range (0,20):
#                for j in range(0,20):
#                    print(mapka.tablica[i][j].cecha, end='')
#                print('')
    def onclick (self, status, x, y, buttons):
        if (status.start==1):
            buttons[self.START.x][self.START.y].config(background="white")
            self.SetStart(wierzcholek (x, y))
            buttons[self.START.x][self.START.y].config(background="green")
        if (status.END==1):
            buttons[self.END.x][self.END.y].config(background="white")
            self.SetEND(wierzcholek(x, y))
            buttons[self.END.x][self.END.y].config(background="red")
        if (status.sciana==1):
            buttons[x][y].config(background="black")
            self.donaldtrump(x, y)
#        self.display()
#        print(self.duzeX)
#        print(self.duzeY)
        
                
    def trasuj(self, buttons):
#        print('a')
        obecny=self.tablica[self.END.x][self.END.y]
        obecny.trasa=1
        self.tablica[obecny.x][obecny.y]=obecny
#        print('b')
        while(obecny.cecha!=0):
            kier=kierunki()
            temp=obecny
            for j in range(0, len(kier.tablica)):
#                print('c')
                duzeX=obecny.x+kier.tablica[j].x
                duzeY=obecny.y+kier.tablica[j].y 
                if((duzeX<0 or duzeX>self.duzeX-1 or duzeY<0 or duzeY>self.duzeY-1)!=1):
                    skan=self.tablica[duzeX][duzeY]
                    if (skan.cecha!=-1 and skan.sciana!=1 and skan.cecha<temp.cecha):
                        temp=skan
            obecny=temp
            obecny.trasa=1
            self.tablica[obecny.x][obecny.y]=obecny
        for x in range(0, self.duzeX):
#            print ("bbbbb")
            for y in range(0, self.duzeY):
#                print("cccccc")
                buttons[x][y].config(text=self.tablica[x][y].cecha)
                if(self.tablica[x][y].trasa):
                    buttons[x][y].config(background="blue")
            
    def trasa(self, buttons):
#        print('a')
        self.CECHUJMAPE(buttons)
#        print('b')
        self.trasuj(buttons)
#        print('c')
#        self.display()
#        print('d')
#        print("aaaa")

        
    
    def display(self):
        for i in range (0,self.duzeX):
            for j in range(0,self.duzeY):
                if(self.tablica[i][j].sciana):
                    print('X', end ='')
                elif(self.tablica[i][j].trasa):
                    print('1',end ='')
                elif(self.tablica[i][j].cecha==0):
                    print('S',end ='')
                elif(i==self.END.x and j==self.END.y ):
                    print('K',end ='')    
                else:
                    print("0", end= '')
            print('')
    def nowywymiar(self,status,buttons):
        X = tkinter.simpledialog.askinteger("Custom size", "Enter amount of rows")
        Y = tkinter.simpledialog.askinteger("Custom size", "Enter amount of columns")
        self=mapa(X,Y)
        restart(self,status,buttons)
#        self.display()
    
window = tkinter.Tk()

window.title("Algorytm dyfuzyjny")    
    
def prepareWindow(mapa,status,buttons):
    buttons=[]
    for x in range(0, mapa.duzeX):
        buttons.append([])
        for y in range(0, mapa.duzeY):
            b = tkinter.Button(window, text=" ", width=2, command=lambda x=x,y=y: mapa.onclick(status,x,y,buttons))
            b.grid(row=x+1, column=y, sticky=tkinter.N+tkinter.W+tkinter.S+tkinter.E)
            buttons[x].append(b)
            buttons[x][y].config(text=mapa.tablica[x][y].cecha, background="white")
            if (mapa.tablica[x][y].sciana):
                buttons[x][y].config(background='black')
#    print('1')
    START=tkinter.Button(window, text="START", command=status.starte,background='green', foreground="white").grid(row=0, column=0, columnspan=3, sticky=tkinter.N+tkinter.W+tkinter.S+tkinter.E)
#    print('2')
    KONIEC=tkinter.Button(window, text="KONIEC", command= status.Kuniec,background='red', foreground="white").grid(row=0, column=3, columnspan=3, sticky=tkinter.N+tkinter.W+tkinter.S+tkinter.E)
#    print('3')
    SCIANA=tkinter.Button(window, text="SCIANA", command= status.BOB,background='black', foreground="white").grid(row=0, column=6, columnspan=3, sticky=tkinter.N+tkinter.W+tkinter.S+tkinter.E)
#    print('4')
    TRASUJ=tkinter.Button(window, text="TRASUJ", command=lambda: mapa.trasa(buttons), background='blue', foreground="white").grid(row=0, column=9, columnspan=3, sticky=tkinter.N+tkinter.W+tkinter.S+tkinter.E)


            
    
    

            
def createMenu(mapa,status,buttons):
    menubar = tkinter.Menu(window)
    menusize = tkinter.Menu(window, tearoff=0)
    menubar.add_command(label="ustaw wymiar", command=lambda: mapa.nowywymiar(status,buttons))
    window.config(menu=menubar)
    
def restart(mapa,status,buttons):
    #destroy all - prevent memory leak
    for x in window.winfo_children():
        if type(x) != tkinter.Menu:
            x.destroy()
    prepareWindow(mapa,status,buttons)   
    
    
class status:
    def __init__(self):
        self.start=0
        self.END=0
        self.sciana=0
    def starte (self):
        self.start=1
        self.END=0
        self.sciana=0
    def BOB (self):
        self.start=0
        self.END=0
        self.sciana=1
    def Kuniec (self):
        self.start=0
        self.END=1
        self.sciana=0
                
mapka= mapa(20,20)
#poprzedniki=grupa()
#poprzedniki.tablica.append(mapka.START)
#print("kllululuuu")
#print(poprzedniki.tablica[0].cecha)
#print("kkukukuu")
#nast=poprzedniki.nastepniki(mapka)
#print(nast.tablica[0].cecha)
#nast.cechujnastepniki(mapka)
#nast2=nast.nastepniki(mapka)
#nast2.cechujnastepniki(mapka)
#mapka.CECHUJMAPE()
#mapka.trasuj()

#for k in range(0, len(nast.tablica)):
#                print(nast.tablica[k].cecha)

    
#for i in range (0,20):
#    for j in range(0,20):
#        if(mapka.tablica[i][j].sciana):
#            print('X',end ='')
#        else:
#            print('O', end= '')
#    print('')
    
#for i in range (0,20):
#    for j in range(0,20):
#        print(mapka.tablica[i][j].cecha, end='')
#    print('')
    
#mapka.display()

buttons=[]

#print(mapka.tablica[5][5].DIST(mapka.tablica[5][6]))
#print(mapka.tablica[5][5].cecha)


stat=status()
#print("dzieki")
createMenu(mapka,stat,buttons)
prepareWindow(mapka,stat,buttons)
#print("dziala")

window.mainloop()
    
```
